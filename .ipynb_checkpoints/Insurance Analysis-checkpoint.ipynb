```python
import sqlite3
import pandas as pd

df = pd.read_csv('insurance.csv')

conn = sqlite3.connect('insurance.db')
#conn.execute('PRAGMA journal_mode=WAL')
#conn.commit()
cursor= conn.cursor()


table_name = 'policy_holders'

columns = ", ".join([f"{col.replace(' ', '_')}" for col in df.columns]) # Eg: page_id TEXT, name TEXT, urslug TEXT, ...
create_table_query = f"CREATE TABLE IF NOT EXISTS {table_name} ({columns});"

cursor.execute(create_table_query)
cursor.fetchall()
```

```python
cursor.execute('pragma table_info(policy_holders);')
cursor.fetchall()
```

```python
# ***TABLE ALREADY EXISTS!! DO NOT RUN THIS CELL!!!!!!***

for index, row in df.iterrows():
    values = ", ".join([f'"{row_item}"' for row_item in row])
    insert_sql = f"INSERT INTO {table_name} ({', '.join(df.columns.str.replace(' ', '_'))}) VALUES ({values})"
    cursor.execute(insert_sql)
    conn.commit()
```

```python
conn = sqlite3.connect('insurance.db')
cursor= conn.cursor()

cursor.execute('SELECT age, sex, bmi, smoker, charges FROM policy_holders')
cursor.fetchall()
```

```python
# CALCULATING AVERAGE BMI FOR MALES WHO ARE SMOKERS
cursor.execute('SELECT AVG(bmi) FROM policy_holders WHERE smoker = "yes" AND sex = "male"')
cursor.fetchone()
```

```python
# CALCULATING AVERAGE BMI FOR MALES WHO ARE NON-SMOKERS
cursor.execute('SELECT AVG(bmi) FROM policy_holders WHERE smoker = "no" AND sex = "male"')
cursor.fetchone()
```
