```python
import selenium.webdriver as driver
from logging import Logger
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Browser Class Simple Version
class Browser:
    
    def __init__(self, browser = driver.Chrome(), address = "https://www.videoschool.com"):
        self.browser = browser
        self.address = address
    
    def goto(self):
        self.browser.get(self.address)
    
    def search(self, method: str, value: str):
        if method == "name":
            self = self.browser.find_element(By.NAME, value)
        if method == "xpath":
            self = self.browser.find_element(By.XPATH, value)
        if method == "css_selector":
            self = self.browser.find_element(By.CSS_SELECTOR, value)
        if method == "class_name":
            self = self.browser.find_element(By.CLASS_NAME, value)
        else:
            raise Exception ("Invalid search method provided")
        return self

    def get_text(self, attribute: str) -> str:
        return self.get_attribute(attribute)
    
    def get_selected_dropdown_text(self) -> str:
        return self.text()
    
    def select_dropdown(self, option: str):
        self.click().search("name", option).click()
    
    def send_keys(self, text: str):
        self.send_keys(text, True)
    
    def clear(self):
        self.clear()

    def click(self):
        self.click()
    
    def close(self):
        self.close()
    
    def end(self):
        self.search.send_keys("end " + Keys.ENTER)
```

```python
# Browser Class Professional Version

import selenium.webdriver as driver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

class Browser:
    
    def __init__(self, browser = driver.Chrome(), address = "https://www.videoschool.com"):
        self.browser = browser
        self.address = address
        self.element = None
    
    def goto(self):
        self.browser.get(self.address)
    
    def search(self, method: str, value: str, wait: int = 10, condition: str = "visible"):
        """Searches for an element on the webpage using Selenium.
        
        Args:
            method: The search method to use ("name","xpath",...)
            value: The value associated with the search method.
            wait: The number of seconds to wait for the element to be found.
            condition: The condition to be met to avoid error ("visible" or "clickable"). """
        condition_checker = {
            "visible": EC.visibility_of_element_located,
            "clickable": EC.element_to_be_clickable
        }
        assert method in ["name", "xpath", "class_name", "css_selector"]
    
        locator = (getattr(By, method.upper()), value)
        try:
            self.element = WebDriverWait(self.browser, wait).until(EC.presence_of_element_located(locator))
        except Exception as err:
            print(self.browser.page_source[:1000])
            raise Exception(f"An error has occurred (element not found with condition {condition} in {wait} seconds): {err}")
        return self

    def get_text(self, attribute: str = "") -> str:
        if not self.element:
            return ""
        if attribute:
            return self.element.get_attribute(attribute)
        else:
            return self.element.text
    
    def get_selected_dropdown_text(self) -> str:
        return self.element.text
    
    def select_dropdown(self, option: str):
        self.click().search("name", option).click()
    
    def send_keys(self, text: str, clear: bool):
        if clear:
            self.element.clear()
            self.element.send_keys(text)
        else:
            self.element.send_keys(text)

    def click(self):
        self.element.click()
    
    def close(self):
        self.browser.close()
    
    def enter(self, value: str):
        self.element.send_keys(value + Keys.ENTER)
```
